```python
# get missing values
df.isnull().sum()
```

```python
# split the dataset
n = len(df)
n_val = int(n * 0.2)

n_test = int(n * 0.2)

n_train = n - n_val - n_test # get remainder because getting 60% with rounding is not accurate

n_val, n_test, n_train
```

```python
n_val + n_test + n_train == n
```

```python
# create new dataframes for each set
df_train = df.iloc[:n_train] # starting from  the beginning to 60% of dataset
df_val = df.iloc[n_train:n_train + n_val] # starting from 60% to 80%
df_test = df.iloc[n_train + n_val:] # starting from 80% to the end
np.arange(n) #  get the range of dataset
```

```python
# shuffle records
idx = np.arange(n)  #  assign var to array of rows (indices)
np.random.seed(2)  #  make the shuffle reproducible
np.random.shuffle(idx) # shuffle the rows

# add in the new shuffled df named idx
df_train = df.iloc[idx[:n_train]] 
df_val = df.iloc[idx[n_train:n_train + n_val]]
df_test = df.iloc[idx[n_train + n_val:]]
```

```python
df_train.head()
```

```python
# make sure the lengths match the originals
n_val, n_test, n_train
```

```python
df_train
```

```python
# reset the indices to start at 0 and drop the index column
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test= df_test.reset_index(drop=True)
```

```python
# apply logarithmic function and get target variables (y)
y_train = np.log1p(df_train.msrp.values)
y_val = np.log1p(df_val.msrp.values)
y_test = np.log1p(df_test.msrp.values)
```

```python
# remove the msrp var from the sets so we don't accidentally use it to train the model
del df_train['msrp']
del df_val['msrp']
del df_test['msrp']
```
